## NYC Crime Data Cleaner Notebook

This notebook cleans the NYC crime data [available from NYC Open Data](#https://data.cityofnewyork.us/Public-Safety/NYPD-Complaint-Data-Historic/qgea-i56i), organizes and saves it to a .csv to be joined to other data sets in the 'Master DataFrame' notebook. 

**Notebook Contents**

> 1. Read in raw CSV selecting relevant columns, cleaning NaN values. 
> 2. Convert DATE column to datetime, slice main DataFrame by date and desired Borough
> 3. Run zonebuilder function to organize lat/lon values into zones (see function docstring for more info)
> 4. Visualize on map, and export to CSV.

```python
import pandas as pd
import numpy as np
import datetime as dt
import zonebuilder as zb
```

## 1. Read in CSV and Cull Nulls

Approximately 7,300,000 rows culled to just under ~7,000,000. Primary culling due to lack of lat/lon values. This could be because the crime was not reported in situ.

The following cells outline the source of the null values in detail.

```python
df = pd.read_csv('./NYPD_Complaint_Data_Historic.csv',usecols = ['BORO_NM','OFNS_DESC','RPT_DT','LAW_CAT_CD','Longitude','Latitude'])
df.info()
```

```python
df[['OFNS_DESC','RPT_DT','LAW_CAT_CD','Longitude','Latitude']].isnull().sum()
```

```python
df.dropna(subset=['Longitude','Latitude'],inplace=True)
df[['OFNS_DESC','RPT_DT','Longitude','Latitude']].isnull().sum()
```

```python
df.dropna(subset = ['OFNS_DESC','RPT_DT'],inplace=True)
df.shape
```

## 2. Slice by Date time frame + Borough

Current slice is by 2016 and only Manhattan. From ~7,000,000 to ~37,500

```python
#convert to date time to slice a particular time frame
df['DATE']=pd.to_datetime(df['RPT_DT'], format = '%m/%d/%Y')
df.drop(['RPT_DT'], axis = 1, inplace=True)
df.info()
```

```python
#TO-DO: Wrap this in a function
sub_df = df[(df['DATE'].dt.date >= pd.to_datetime('01/01/2014', format ='%m/%d/%Y')) & (df['DATE'].dt.date <= pd.to_datetime('12/31/2015', format ='%m/%d/%Y')) & (df['BORO_NM'] == 'MANHATTAN')].copy()
```

```python
sub_df.drop(['BORO_NM'],axis=1,inplace=True)
sub_df.info()
```

## 3. Bin Data into Zones Based on Latitude + Longitude

The binning process has been modified from a script written to create a grid of 'zones' over Manhattan (zonebuilder.py in project folder), to a much quicker and more generalized approach of simply rounding down the lat/lon numbers to the nearest thousandth. 

This creates a 'grid' of approximately 365.23FT East/West and 276.94FT North/South that all data points will fall into. 

The reasoning for this came about by looking at a map of the values plotted and noticing an emergent grid for each block. It looks like corner/mid block are hard coded 'options' when a police officer is logging a complaint. So despite the apparent high degree of precision seen in the DataFrame, there is a subset of unique combinations. 

see raw plot below (data straight onto a map) + modified set to generalize across data sets (long/lat rounded to 3 decimal points)

```python
sub_df.head()
```

```python
sub_df['Coords'] = list(zip(sub_df.Longitude, sub_df.Latitude))
```

```python
rnd_df = pd.DataFrame()
rnd_df = sub_df
rnd_df = rnd_df.join(pd.get_dummies(sub_df['LAW_CAT_CD']))
rnd_df[['Latitude','Longitude']]=rnd_df[['Latitude','Longitude']].round(2)
rnd_df['Coords'] = list(zip(rnd_df.Longitude, rnd_df.Latitude))
rnd_df.head()
```

```python
rnd_df.info()
```

## 4. Plotting to Map,  EDA

rnd_df goes to csv, create a new df to aggregate sample data. Sanity checks, etc.

Below, find map plotted showing bins with crime values

```python
rnd_df.to_csv('20142015crime.csv')
```

```python
rnd_df.describe()
```

```python
g_rnd_df = pd.DataFrame()
g_rnd_df = rnd_df.groupby(['Coords']).sum()
g_rnd_df = g_rnd_df.reset_index()
g_rnd_df['Longitude'], g_rnd_df['Latitude'] = zip(*g_rnd_df.Coords)
```

```python
g_rnd_df.describe()
```

```python
import folium
from IPython.display import IFrame

nyc_map = folium.Map([40.78, -73.97], tiles = 'CartoDB positron')

for lat, long, fel, mis, vio in zip(g_rnd_df['Latitude'], g_rnd_df['Longitude'], 
                                         g_rnd_df.FELONY, g_rnd_df.MISDEMEANOR, g_rnd_df.VIOLATION):

    folium.Marker( [lat, long], 
                       icon=folium.CustomIcon(icon_image='https://i.imgur.com/CYx04oC.png',
                                              icon_size=(10,10) ), popup='Felony '+str(fel)+'\n\n'+'Misdem '+str(mis)+'\n\n'+'Violation '+str(vio)).add_to(nyc_map)

nyc_map.save('nyc_crime.html')
IFrame(src='nyc_crime.html', width='100%', height=500)
```
